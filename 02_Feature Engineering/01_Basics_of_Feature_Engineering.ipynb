```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
import pandas as pd
import numpy as np

# Loading Titanic Dataset
df = pd.read_csv(url)

# Displaying Information
print(df.info())
```

```python
print("--- Dataset Preview ---\n", df.head(5))
```

```python
# Seperate Featrues
categorical_features = df.select_dtypes(include=['object']).columns
numerical_features = df.select_dtypes(include=["float64", "int64"]).columns

print("\nCategorical Features: \n", categorical_features.tolist())
print("\nNumerical Feature: \n", numerical_features.tolist())
```

```python
# Display Summary of Categorical Data
print("\nCategorical Data Summary: \n")
for col in categorical_features:
    print(f"{col}:\n", df[col].value_counts(), "\n")

print("\nNumerical Data Summary :\n")
print(df[numerical_features].describe())
```
