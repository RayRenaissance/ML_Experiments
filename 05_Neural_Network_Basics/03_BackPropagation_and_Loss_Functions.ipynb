```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Mean Squared Error (MSE) Loss
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

# Binary Cross Entropy (BCE) loss
def binary_cross_entropy(y_true, y_pred):
    y_pred = np.clip(y_pred, 1e-15, 1 - 1e-15)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
```

```python
# Example Dataset
y_true = np.array([1, 0, 1, 1, 0, 1])
y_pred = np.array([0.9, 0.2, 0.7, 0.6, 0.4, 0.8])

# Calculate Loss
mse = mse_loss(y_true, y_pred)
bce = binary_cross_entropy(y_true, y_pred)
```

```python
# Display Calculated Loss
print(f"Mean Squared Error LOSS: {mse:.4f}")
print(f"Binary Cross-Entropy LOSS: {bce:.4f}")
```

```python
# Derivative of MSE Loss 
def mse_gradients(y_true, y_pred):
    return 2 * (y_pred - y_true) / len(y_true)

# Derivative of BCE Loss
def bce_gradients(y_true, y_pred):
    y_pred = np.clip(y_pred, 1e-15, 1 - 1e-15)
    return (y_pred - y_true) / (y_pred * (1 - y_pred))
```

```python
# Calculate the Gradients and Display 
grad_mse = mse_gradients(y_true, y_pred)
grad_bce = bce_gradients(y_true, y_pred)
print(f"Mean Squared Error Gradients: \n{grad_mse}")
print(f"Binary Cross-Entropy Gradients: \n{grad_bce}")
```

```python
# Visualization: Define Prediction and True Label
predictions = np.linspace(0, 1, 100)
true_label = 1

# Compute Losses
mse_losses = [(true_label - p)**2 for p in predictions]
bce_losses = [-true_label * np.log(max(p, 1e-15)) - (1 - true_label) * np.log(max(1 - p, 1e-15)) for p in predictions]

plt.figure(figsize=(10, 6))
plt.plot(predictions, mse_losses, label="Mean Squared Error Losses")
plt.plot(predictions, bce_losses, label="Binary Cross-Entropy Losses")
plt.title("Loss Functions Comparison")
plt.xlabel("Predictions")
plt.ylabel("Losses")
plt.grid(True)
plt.legend()
plt.show()
```
