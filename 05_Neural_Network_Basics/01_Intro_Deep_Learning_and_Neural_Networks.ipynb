```python
# Load and Explore MNIST and CIFAR10 
from tensorflow.keras.datasets import mnist, cifar10
import tensorflow as tf
import torch.nn as nn
import torch
import matplotlib.pyplot as plt

# MNIST
(X_train_mnist, y_train_mnist), (X_test_mnist, y_test_mnist) = mnist.load_data()
print(f"MNIST Dataset: Train = {X_train_mnist.shape}, Test = {X_test_mnist.shape}")

# CIFAR10
(X_train_cifar, y_train_cifar), (X_test_cifar, y_test_cifar) = cifar10.load_data()
print(f"CIFAR-10 Dataset: Train = {X_train_cifar.shape}, Test = {X_test_cifar.shape}")
```

```python
# Define a Basic Dense Layer using Tensorflow
tf_layer = tf.keras.layers.Dense(units=10, activation='relu')
print(f"Tensorflow Layer: {tf_layer}")

# Define a Basic Dense Layer using PyTorch
nn_layer = nn.Linear(in_features=10, out_features=5)
print(f"PyTorch Layer: {nn_layer}")
```

```python
# Visualize MNIST Sample
plt.imshow(X_train_mnist[0], cmap='gray')
plt.title(f"MNIST Label: {y_train_mnist[0]}")
plt.show()

# Visualize CIFAR10 Sample
plt.imshow(X_train_cifar[0], cmap='gray')
plt.title(f"MNIST Label: {y_train_cifar[0]}")
plt.show()
```
