### Import and Define Target and Variable

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)
df.head(1)
```

```python
# Defining Feature and Target 
features = df[["total_bill", "size"]]
target = df["tip"]

# Checing Feature and Target 
print("\n--- Features ---\n", features.head(3))
print("\n--- Target ---\n", target.head(3))
```

```python
# Splitting Data into Training and Testing
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

print("--- Training Data Set ---\n", X_train.shape)
print("\n--- Testing Data Set ---\n", X_test.shape)
```

```python
# Visualize The Data
sns.pairplot(df, x_vars=["total_bill", "size"], y_vars="tip", height=5, aspect=0.8, kind="scatter")
plt.title("Features vs Target Relationship")
plt.show()
```
